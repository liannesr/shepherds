```python
#hide
!pip install -Uqq fastbook
import fastbook
fastbook.setup_book()

#hide
# LSR: Import everything from fastbook and the vision widgets from fastai library
from fastbook import *
from fastai.vision.widgets import *
```

```python
path = Path()
```

```python
learn_inf = load_learner(path/'export.pkl', cpu=True)
```

```python
btn_upload = widgets.FileUpload()
```

```python
out_pl = widgets.Output()
```

```python
lbl_pred = widgets.Label()
```

```python
# LSR: Create the button upload widget again.
#hide
#Putting back btn_upload to a widget for next cell
btn_upload = widgets.FileUpload()

btn_upload
```

```python
# LSR: We create an image from the uploaded.
img = PILImage.create(btn_upload.data[-1])
```

```python
# LSR: Here we create a button so we can activate the classification process. 
#hide_output
btn_run = widgets.Button(description='Classify')
```

```python
# LSR: The event handler is created to service everytime the button is pressed.
def on_click_classify(change):
    # LSR: Create the image from the upload widget.
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    # LSR: Display the image thumbnail that we got from the widget.
    with out_pl: display(img.to_thumb(128,128))
    # LSR: Perform the prediction on the widget saving prediction and probabilities.
    pred,pred_idx,probs = learn_inf.predict(img)
    # LSR: Save the values. 
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

# LSR: Attach to the button the event listener.
btn_run.on_click(on_click_classify)
```

```python
# LSR: We show as output the image that I uploaded (teddy bear instead of the grizzly).
#hide_output
out_pl = widgets.Output()
out_pl.clear_output()

# LSR: For each element in the widget make a thumbnail of size 128 x 128.
with out_pl: display(img.to_thumb(128,128))
    
# LSR: Display the thumbnail. 
out_pl
```

```python
# LSR: Call predict to make predictions on the chosen image.
pred,pred_idx,probs = learn_inf.predict(img)

# LSR: He use the label widget to get the what was the prediction and the probability of it. 
#hide_output
lbl_pred = widgets.Label()
lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

# LSR: Show the prediction in the console. 
lbl_pred
```

```python
# LSR: Make the event listener tied to buttons, uploads, output, and predictions.
# LSR: Puts every item in a vertical box.
# LSR: Tried it with a pink bear; Classified correctly with 99% confidence.
#hide_output
VBox([widgets.Label('Select your bear!'),
      btn_upload, btn_run, out_pl, lbl_pred])
```
